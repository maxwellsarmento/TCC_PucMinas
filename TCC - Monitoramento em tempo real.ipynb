```python
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.animation import FuncAnimation
%matplotlib notebook
```

```python
plt.style.use('fivethirtyeight')
```

```python
frame_len = 1000

fig = plt.figure(figsize=(9,6))

def animate(i):
    data = pd.read_csv('aneel_distribuidoras.csv')
    y1 = data['Light']
    y2 = data['Grupo Enel']

    if len(y1)<=frame_len:
        plt.cla()
        plt.plot(y1, label='Light')
        plt.plot(y2, label='Grupo Enel')
    else:
        plt.cla()
        plt.plot(y1[-frame_len: ], label='Light')
        plt.plot(y2[-frame_len: ], label='Grupo Enel')
    
    plt.legend(loc='upper left')
    plt.tight_layout()


ani = FuncAnimation(plt.gcf(), animate, interval=2000)
```
